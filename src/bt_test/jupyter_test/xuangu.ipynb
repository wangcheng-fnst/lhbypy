```python
import tushare as tu
import pandas as pd
import talib as ta


```

```python
stock_codes = pd.read_csv('stock.csv')
stock_codes = stock_codes['code']
```

```python
def choose(df):
    stock_data = df
    stock_data['ma5'] = ta.EMA(stock_data['close'], timeperiod=5)
    stock_data['ma10'] = ta.EMA(stock_data['close'], timeperiod=10)
    stock_data['ma30'] = ta.EMA(stock_data['close'], timeperiod=30)
    stock_data['ma250'] = ta.EMA(stock_data['close'], timeperiod=250)
    hit = (stock_data['ma5'][-1] >= stock_data['ma10'][-1])  \
        and (stock_data['ma10'][-1] >= stock_data['ma30'][-1]) \
        and (stock_data['ma30'][-1] >= stock_data['ma250'][-1])
    return hit
```

```python
def get_data(code):
    df = tu.get_k_data(code=code)
    if not df.empty:
        df.index = pd.to_datetime(df['date'])
        return df[['open','close','high','low','volume']]
    return None

hit_codes =[]

for code in stock_codes[1000:1050]:
    df = get_data(str(code))
    if df is not None:
        if choose(df):
            hit_codes.append(code)

print ('all codes: %s'% hit_codes)
```
