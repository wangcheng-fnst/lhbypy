```python
def get_data(code):
    df = tu.get_k_data(code=code)
    if not df.empty:
        df.index = pd.to_datetime(df['date'])
        return df[['open','close','high','low','volume']]
    return None

df = get_data('002104')
df
```

```python
df['up'] = (df['close'] - df.shift(1)['close']) / df.shift(1)['close'] >= 0.099
df[df['up']]
```
